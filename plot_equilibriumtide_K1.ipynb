```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from scipy.io import loadmat
import basemap_func as bm
```

# Plot equilibrium tide M2

### Define constants and arrays

```python
# Set amplitdue A and Love number combination 1+k2-h2
A = 14.1565/100 # units: m
Lovenumbercombo = 0.736

# Create 1-d vectors of lat and lon
latsvector = np.arange(-90,90)
lonsvector = np.arange(360)

# Create 2-d arrays of lat and lon values
lonsarray,latsarray = np.meshgrid(lonsvector,latsvector)
```

### Compute and plot equilibrium tide for omega*t = 0 and omega*t=180

```python
# Set latitudinal and longitudinal dependence
latsdependence = np.sin(2 * np.deg2rad(latsarray))
lonsdependence = 1

equilibriumtide0 = A * Lovenumbercombo * latsdependence * np.cos(np.deg2rad(0+lonsdependence*lonsarray))
equilibriumtide180 = A * Lovenumbercombo * latsdependence * np.cos(np.deg2rad(180+lonsdependence*lonsarray))
```

### Plot

```python
# Set the default fontsize
font = {'size'   : 18}
matplotlib.rc('font', **font)

plt.figure(figsize=(12,12))
plt.subplot(211)
plt.pcolor(lonsarray,latsarray,equilibriumtide0)
plt.gca().invert_yaxis()
bm.basemap_func()
plt.xlim(0,360)
plt.ylim(-90,90)
plt.ylabel('Degrees Latitude')
plt.title('Snapshot of K1 equilibrium tide in m')
plt.colorbar()

plt.subplot(212)
plt.pcolor(lonsarray,latsarray,equilibriumtide180)
plt.gca().invert_yaxis()
bm.basemap_func()
plt.xlim(0,360)
plt.ylim(-90,90)
plt.xlabel('Degrees Longitude')
plt.ylabel('Degrees Latitude')
plt.title('Snapshot of K1 equilibrium tide in m, 180$^{\circ}$ later')
plt.colorbar()

plt.savefig('Snapshot_K1_equilibriumTide.png')
```
