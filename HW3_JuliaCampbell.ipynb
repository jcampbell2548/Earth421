```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
from mpl_toolkits.basemap import Basemap
```

```python
omega = (2 * np.pi)/ 86400  # converted 1 day to 86400 seconds
theta = range(-90,90,1) # latitudinal range of interest
f = 2 * omega * np.sin(np.deg2rad(theta)) # formula for f from lecture notes, latitude in radians
plt.figure(figsize=(12,8))
plt.plot(theta, f)
plt.xlabel('Latitude (degrees)')
plt.ylabel('f (1/seconds)')
plt.title('Coriolis Parameter f')
```
